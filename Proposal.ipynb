# 2. Webscraping 

## 2.1. URLs Prepration

By considering the data from Box Office Mojo, which is organized on a yearly basis, our objective is to create a list of the specific URLs we intend to use.

```python
URL_list = []

#adding movies' URl from 2000 to 2023 in our list
years = list(range (2000, 2024))
for i in range (len(years)):
    URL = 'https://www.boxofficemojo.com/year/'+ str(years[i]) + '/?grossesOption=calendarGrosses'
    URL_list.append(URL)

URL_list
```

## 2.2. Importing BeutifulSoup and requests

We need to import the `BeutifulSoup` library to scrap information from web pages.
To work with BeutifulSoup We need some HTML or XML as BeutifulSoup sits atop HTML or XML parser to provide Pythonic idioms for iterating, searching, and modifying the parse tree. We can either use a local HTML/XML file, or it can be requested from a web server.

To use the requsted method (online), we have to import another library called `requests`, which is also not built in. The library provides a simple API for interacting with HTTP operations.

```python
from bs4 import BeautifulSoup
import requests
```

## 2.3. Obtaining Responses from HTTP Requests

We have to use the `get` method from the `requests` library to retrieve an HTTP response object. An HTTP request contains header fields which are a list of strings sent and received by both the client program and server on every HTTP request and response. These headers may give the server some additional info about the request.

One of these fields is called: "user-agent" which tells the server the software that made the request in place of the user.
Some websites might block access without using "user-agent".

In order to present if we are getting the proper answer from our website or not we will use the `urllib.error` library.

```python
from urllib.error import HTTPError

#defining a user-agent for our request
HEADERS = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/118.0.0.0 Safari/537.36'
}

#check if we are getting the proper HTTP response status code
data_list = []
for URL in range (len(URL_list)):
    try:
        r = requests.get(URL_list[URL], headers=HEADERS)
        data =r.text
        data_list.append(data)
        r.raise_for_status()
    except HTTPError as hp:
        print (hp)
    else:
        print("it's working")
```

## 2.4. Making the Soup

The raw HTML is not very easy to work with because it is a semantic markup format. Therefore, we parse the HTML and create an object named "soup" to store the parsed HTML content. Additionally, we create a dictionary to associate each of the HTML parses with their respective year.

```python
#creat a list containes "soup+year(2000-2023)" as name for our dictionary keys
soup_name_list = []
for i in range(2000, 2024):
    soup_name = "soup%d" % i
    soup_name_list.append(soup_name)

#parse the raw HTMLs for each webpage
soups_dict = {}
for j in range(len(data_list)):
        soups_dict[soup_name_list[j]] = BeautifulSoup(data_list[j],'html.parser')

soups_dict['soup2000']
```

## 2.5. Extracting elements by tags

Once our soups have been prepared, we will proceed to strain each soup and extract the desired components. Since our data is structured in a table and we require all the data, we will employ the `find_all` method. However, if we have specific preferences for items within our HTML section, we can specify the attributes we are seeking.

```python
# extracting 'td' elements for each webpage 
for j in(soup_name_list):
    s = []
    for i in soups_dict[j].find_all('td'):
        s.append(i.get_text())
    soups_dict.update({j: s})
    
soups_dict['soup2000']
```

## 2.6. Creating our Dataframe 

First of all,to work with dataframes we have to import the `pandas` library. We are planning to create a dictionary (`dict`) for each webpage, where we will store the data from each webpage as separate dataframes. The reason for not summing all the extracted elements and creating a dataframe immediately is that we intend to make individual changes to each year's data. For instance, in the case of `'soup2000'` there are recurring `'-'` symbols in our data that we aim to address. While some of these `'-'` symbols appear after the movie names on the webpage, others indicate an unknown distributor. Instead of deleting these symbols, we opt to replace them with a specific string, such as 'Unknown,' and subsequently exclude them when constructing our dataframe by segmenting the list of extracted data (with every 11 elements representing a column).

Once we have assigned names to each column, we will apply a function to convert them into strings using `str` and then use `strip` to remove any `'\n'` characters that appeared in certain "distributor" columns. Afterward, we will replace all `'-'` symbols with 'Unknown'.

```python
import pandas as pd

# Creating a dataframe for each extracted webpage's data
df_dict = {}
for soup in soup_name_list:
    df_dict[soup] = pd.DataFrame(list(zip(soups_dict[soup][:: 11], soups_dict[soup][1:: 11],
                               soups_dict[soup][5:: 11], soups_dict[soup][6:: 11], soups_dict[soup][7:: 11], soups_dict[soup][9:: 11])), 
                      columns=['Rank', 'Movies', 'gross', 'theater sale', 'total gross','distributor'])
    #removing \n and replacing '-' with 'Unkwon'
    df_dict[soup]= df_dict[soup].apply(lambda x: x.str.strip()).replace('-', 'Unkwon')


df_dict['soup2023']
```

## 2.7. Appending the Year in the Dataframe

In this section, we will introduce a new column named "Year" to display the release year of each movie in every dataframe. To achieve this, we have developed a dictionary where the dataframe names (representing each year) are the keys, and their corresponding release years are the values. Thereafter, we will apply this column across all the dataframes through an iterative process.

```python
#creating a dictionary to link each dataframe with the year of release
year_mapping_dict = {
    'soup2000': 2000,'soup2001': 2001,'soup2002': 2002,'soup2003': 2003,'soup2004': 2004,'soup2005': 2005,
    'soup2006': 2006,'soup2007': 2007,'soup2008': 2008,'soup2009': 2009,'soup2010': 2010,'soup2011': 2011,
    'soup2012': 2012,'soup2013': 2013,'soup2014': 2014,'soup2015': 2015,'soup2016': 2016,'soup2017': 2017,
    'soup2018': 2018,'soup2019': 2019,'soup2020': 2020,'soup2021': 2021,'soup2022': 2022,'soup2023': 2023
}

#assigning the release year to each element in our dataframe
for i in soup_name_list:
    if i in year_mapping_dict:
        df_dict[i]['Year'] = year_mapping_dict[i]


df_dict['soup2023']
```

## 2.8 Modifying elements

### 2.8.1. String to integer

To proceed with the analysis of our data concerning the total gross, it is necessary to convert the data type of this column from a string to either an integer or a float. To achieve this, we need to eliminate the "$" and "," symbols from each element within the "total gross" column.

```python
#removing '$' and ',' from "total gross" coloumn and making "total gorss" values integer
for j in soup_name_list:
    for i in range(len(df_dict[j])):
        total_gross_value = str(df_dict[j]['total gross'][i])
        theater_value = str(df_dict[j]['theater sale'][i])
        total_gross_value = int(total_gross_value.replace('$', '').replace(',', ''))
        #some values in "theater sale" column is Unkown
        try:
            theater_value = int(theater_value.replace(',', ''))
        except:
            pass
        df_dict[j].at[i, 'total gross'] = total_gross_value
        df_dict[j].at[i, 'theater sale'] = theater_value

        
df_dict['soup2023']
```

### 2.8.3. Considering infelation and Saving our data

As inflation changes the actual value of the dollar over time, I believe it is inequitable to directly compare a movie from 2000 with one from 2023. Therefore, we intend to adjust the total gross value of movies by multiplying an inflation rate.

We'll obtain the inflation rate from the website: https://www.in2013dollars.com/, using the 2023 dollar as the baseline value against which all other values will be compared.

To protect against potential changes in online data or server disruptions from our primary data source, we will extract and store each dataframe on our own machine to avoid any potential loss.

```python
#creating a dictionary to link inflation factor with the year of release
inflation_mapping_dict = {
    'soup2000': 1.79,'soup2001': 1.74,'soup2002': 1.71,'soup2003': 1.67,'soup2004': 1.63,'soup2005': 1.58,
    'soup2006': 1.53,'soup2007': 1.48,'soup2008': 1.43,'soup2009': 1.43,'soup2010': 1.41,'soup2011': 1.37,
    'soup2012': 1.34,'soup2013': 1.32,'soup2014': 1.30,'soup2015': 1.30,'soup2016': 1.28,'soup2017': 1.26,
    'soup2018': 1.23,'soup2019': 1.20,'soup2020': 1.19,'soup2021': 1.14,'soup2022': 1.05,'soup2023': 1.00
}

#multiplying the inflation factor by each element of the "total gross" in each year
for i in soup_name_list:
    if i in inflation_mapping_dict:
        df_dict[i]['total gross'] = (df_dict[i]['total gross']*inflation_mapping_dict[i])
        try:
            df_dict[i]['theater sale'] = (df_dict[i]['theater sale']*inflation_mapping_dict[i])
        except:
            pass
#saving each dataframe into our machine
    df_dict[i].to_csv('df%s.csv' %i, sep=',', index =False, encoding='utf-8')


df_dict['soup2000']        
```

```python
#Boxplot
years = list(range (2000, 2024))

plt.figure(figsize=(30, 20))

count = 1
for i in soup_name_list:
    bx = plt.gca().boxplot(df_dict[i]['total gross'], positions=[count], vert=True, patch_artist=True)
    count += 1

plt.xticks(range(1, len(years) + 1), [f'{year} Movies' for year in years])
plt.xlabel('Movies')
plt.ylabel('Total Gross')
plt.title('Boxplot of Total Gross')

plt.xticks(rotation=45)
plt.show()
```
